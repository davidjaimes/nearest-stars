```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# REad exoplanet encyclopedia data.
fname = '../data/exoplanet.eu_catalog_2021-05-22.csv'
columns = ['star_name', 'ra', 'dec', 'star_distance',
           'star_radius', 'star_teff', 'mag_v']
df = pd.read_csv(fname, usecols=columns)
df = df.drop_duplicates()

# Convert visual mag to absolute mag.
df['Absmag'] = df['mag_v'] - 5 * (np.log10(df['star_distance']) - 1)

# Convert parsecs to light year.
df['star_distance'] *= 3.2616
w = df['star_distance'] <= 50
df = df[w]
len(df)
```

```python
plt.scatter(df['star_teff'], df['Absmag'],
            s=100*df['star_radius'], c=df['star_distance'])
plt.xlim(10000, 500)
plt.ylim(20, 0)
```
